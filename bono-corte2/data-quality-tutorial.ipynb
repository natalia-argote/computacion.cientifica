<span style="color:red">TODO: Create a function to fix the digitation errors for column "DIA". Make sure to update original dataframe.</span>

```python
comparision_array = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
for dia in homicides_df['DIA']:  
    homicides_df["DIA"] = fix_errors(dia)

def fix_errors(name):
    day = ''
    cont = len(name)
    for c in comparision_array:
        distance_words = pylev.levenshtein(c, name)
        if(distance_words < cont):
            day = c
            cont = distance_words
    return day

homicides_df.head()

# Showing different values for "DIA" column
homicides_df["DIA"].unique()
```
